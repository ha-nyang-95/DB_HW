```python
# 3. 결측치 확인
# 각 컬럼별 결측치 개수를 확인
print("\n각 컬럼별 결측치 개수")
df.isnull().sum()  # 결측치를 확인하는 함수 (힌트: 결측치를 확인하는 함수는 isna() 또는 isnull())
```

```python
# 4. 결측치 처리
# (1) Age 컬럼의 결측치를 중앙값(median)으로 대체
age_median = df["Age"].median()  # 중앙값을 구하는 함수 (힌트: 중앙값을 계산하는 함수는 median())
df["Age"].fillna(age_median, inplace=True)  # 결측치를 중앙값으로 채우는 함수 (힌트: fillna 사용)

# (2) Embarked 컬럼의 결측치를 최빈값(mode)으로 대체
embarked_mode = df["Embarked"].mode()[0]  # 최빈값을 구하는 함수 (힌트: 최빈값을 찾는 함수는 mode())
df["Embarked"].fillna(embarked_mode, inplace=True)  # 결측치를 최빈값으로 채우는 함수 (힌트: fillna 사용)

# 5. 결측치 처리 후 다시 확인
print("\n결측치 처리 후 각 컬럼별 결측치 개수:")
print(df.isnull().sum())  # 결측치가 잘 처리되었는지 확인하는 함수 (힌트: isna() 또는 isnull() 사용)
```

```python
# 6. 주요 통계 정보 확인
# 숫자형 데이터의 기초 통계 확인 (평균, 표준편차, 최소/최대값 등)
print("\n숫자형 데이터 요약 통계")
df.describe()  # 숫자형 데이터의 통계 요약을 출력하는 함수 (힌트: describe())
```

```python
# 7. 범주형 데이터(객실 등급, 성별, 탑승항) 값 분포 확인
print("\n객실 등급(Pclass)별 승객 수")
print(df["Pclass"].value_counts())  # Pclass 값 개수를 세는 함수 (힌트: value_counts())

print("\n성별(Sex)별 승객 수:")
print(df["Sex"].value_counts())  # Sex 값 개수를 세는 함수 (힌트: value_counts())

print("\n탑승 항구(Embarked)별 승객 수:")
print(df["Embarked"].value_counts())  # Embarked 값 개수를 세는 함수 (힌트: value_counts())
```

```python
import pandas as pd  # Pandas 라이브러리 불러오기

# 1. Titanic 데이터셋 불러오기 (파일 경로는 사용자 환경에 맞게 변경)
df = pd.read_csv("titanic.csv")  # Titanic 데이터셋 파일명을 입력하세요.

# 2. 데이터프레임 기본 정보 확인
print("데이터프레임 기본 정보")
print(df.info())  # 데이터프레임의 전체 구조를 확인하는 함수 입력
```

```python
# 3. 특정 조건에 따른 데이터 필터링

# (1) 1등석 승객만 필터링
first_class_passengers = df[df["Pclass"] == 1]  # 1등석 승객만 필터링하는 조건을 입력하세요.
print("\n1등석 승객 데이터")
first_class_passengers.head()  # 상위 5개 데이터를 확인하는 함수 입력
```

```python
# (2) 여성 승객만 필터링
female_passengers = df[df["Sex"] == "female"]  # 여성 승객만 필터링하는 조건을 입력하세요.
print("\n여성 승객 데이터")
female_passengers.head()  # 상위 5개 데이터를 확인하는 함수 입력
```

```python
# (3) 30세 이상 승객만 필터링
adult_passengers = df[df["Age"] >= 30]  # 30세 이상 승객만 필터링하는 조건을 입력하세요.
print("\n30세 이상 승객 데이터")
adult_passengers.head()  # 상위 5개 데이터를 확인하는 함수 입력
```

```python
# (4) 3등석에 탑승한 20세 이하 여성 승객 필터링
young_female_3rd_class = df[(df["Pclass"] == 3) & (df["Sex"] == "female") & (df["Age"] <= 20)]  
# 3등석에 탑승한 20세 이하 여성 승객을 필터링하는 조건을 입력하세요.

print("\n3등석에 탑승한 20세 이하 여성 승객")
young_female_3rd_class.head()  # 상위 5개 데이터를 확인하는 함수 입력
```

```python
# (5) 50세 이상 생존한 승객만 필터링
survived_senior_passengers = df[(df["Age"] >= 50) & (df["Survived"] == 1)]  
# 50세 이상이며 생존한 승객만 필터링하는 조건을 입력하세요.

print("\n50세 이상 생존한 승객")
survived_senior_passengers.head()  # 상위 5개 데이터를 확인하는 함수 입력
```

```python
# (6) 2등석 또는 3등석에 탑승한 남성 승객 필터링
second_or_third_class_male = df[(df["Pclass"].isin([2, 3])) & (df["Sex"] == "male")]  
# 2등석 또는 3등석에 탑승한 남성 승객을 필터링하는 조건을 입력하세요.

print("\n2등석 또는 3등석에 탑승한 남성 승객")
second_or_third_class_male.head()  # 상위 5개 데이터를 확인하는 함수 입력
```

```python
import pandas as pd  # Pandas 라이브러리 불러오기
# =============================================================================
# fillna() 함수 설명
# fillna() 함수는 데이터프레임에서 결측치(NaN)를 특정 값으로 대체하는 데 사용됩니다.
# 평균(mean), 중앙값(median), 최빈값(mode) 또는 고정된 값을 활용하여 결측치를 처리할 수 있습니다.

# 기본 문법:
# df["컬럼명"].fillna(대체할 값, inplace=True)
# =============================================================================
# =============================================================================
# quantile() 함수 설명
# quantile() 함수는 데이터에서 특정 분위수(퍼센트) 값을 구하는 데 사용됩니다.
# 예를 들어, quantile(0.90)은 상위 10%에 해당하는 값의 기준을 의미합니다.

# 기본 문법:
# df["컬럼명"].quantile(퍼센트)
# =============================================================================

# 1. Titanic 데이터셋 불러오기 (파일 경로는 사용자 환경에 맞게 변경)
df = pd.read_csv("titanic.csv")  # Titanic 데이터셋 파일명을 입력하세요.

# 2. 데이터프레임 기본 정보 확인
print("데이터프레임 기본 정보")
df.info()  # 데이터 구조 및 결측치 개수를 확인하는 함수 입력
```

```python
# 3. 생존자 데이터 필터링
# Survived 값이 1인 승객만 필터링 (생존한 승객)
survived_passengers = df[df["Survived"] == 1]  # 생존자만 필터링하는 조건을 입력하세요.
print("\n생존한 승객 데이터 (상위 5개)")
survived_passengers.head()  # 상위 5개 데이터를 확인하는 함수 입력
```

```python
# 4. 사망자 데이터 필터링
# Survived 값이 0인 승객만 필터링 (사망한 승객)
deceased_passengers = df[df["Survived"] == 0]  # 사망자만 필터링하는 조건을 입력하세요.
print("\n사망한 승객 데이터 (상위 5개)")
deceased_passengers.head()  # 상위 5개 데이터를 확인하는 함수 입력
```

```python
# 5. 요금(Fare) 결측치 확인 및 처리
# Fare 컬럼의 결측치 개수 확인
print("\n요금(Fare) 컬럼 결측치 개수")
df["Fare"].isna().sum()  # 결측치 개수를 확인하는 함수 입력
```

```python
# 결측치가 있으면 0으로 채우기
df["Fare"].fillna(0, inplace=True)  # 결측치를 0으로 대체하는 함수 입력
print("\n요금(Fare) 컬럼 결측치 개수")
df["Fare"].isna().sum()  # 결측치 개수를 확인하는 함수 입력
```

```python
# 6. 고액 요금 승객 필터링
# 요금이 상위 10%에 해당하는 값 찾기
fare_threshold = df["Fare"].quantile(0.90)  # 상위 10% 기준값을 구하는 함수 입력

# 고액 요금 승객 필터링 (요금이 fare_threshold 이상인 승객)
high_fare_passengers = df[df["Fare"] >= fare_threshold]  # 고액 요금 승객을 필터링하는 조건을 입력하세요.
print(f"\n고액 요금 승객 데이터 (상위 5개) - 기준 요금: {fare_threshold:.2f}")
high_fare_passengers.head()  # 상위 5개 데이터를 확인하는 함수 입력
```

```python
# 7. 생존한 고액 요금 승객 필터링
# 생존자 중에서 요금이 fare_threshold 이상인 승객
high_fare_survivors = df[(df["Survived"] == 1) & (df["Fare"] >= fare_threshold)]  
# 생존자 중에서 고액 요금 승객을 필터링하는 조건을 입력하세요.

print("\n생존한 고액 요금 승객 데이터 (상위 5개)")
high_fare_survivors.head()  # 상위 5개 데이터를 확인하는 함수 입력
```

```python
# 8. 사망한 고액 요금 승객 필터링
# 사망자 중에서 요금이 fare_threshold 이상인 승객
high_fare_deceased = df[(df["Survived"] == 0) & (df["Fare"] >= fare_threshold)]  
# 사망자 중에서 고액 요금 승객을 필터링하는 조건을 입력하세요.

print("\n사망한 고액 요금 승객 데이터 (상위 5개)")
high_fare_deceased.head()  # 상위 5개 데이터를 확인하는 함수 입력
```

```python
# =============================================================================
# fillna() 함수 설명
# fillna() 함수는 데이터프레임에서 결측치(NaN)를 특정 값으로 대체하는 데 사용됩니다.
# 평균(mean), 중앙값(median), 최빈값(mode) 또는 고정된 값을 활용하여 결측치를 처리할 수 있습니다.

# 기본 문법:
# df["컬럼명"].fillna(대체할 값, inplace=True)
# =============================================================================
# =============================================================================
# value_counts() 함수 설명
# value_counts() 함수는 특정 컬럼에서 고유한 값의 개수를 계산하는 데 사용됩니다.
# 주로 범주형 데이터에서 각 항목이 몇 번 나타났는지 확인할 때 활용됩니다.

# 기본 문법:
# df["컬럼명"].value_counts()
# =============================================================================
# =============================================================================
# isna() 함수 설명
# isna() 함수는 데이터프레임에서 결측치(NaN)를 찾는 데 사용됩니다.
# 결과는 True(결측치) 또는 False(결측치 없음)로 표시됩니다.

# 기본 문법:
# df["컬럼명"].isna()  # 특정 컬럼의 결측치를 확인
# df["컬럼명"].isna().sum()  # 결측치 개수 계산
# =============================================================================

import pandas as pd  # Pandas 라이브러리 불러오기

# 1. Titanic 데이터셋 불러오기 (파일 경로는 사용자 환경에 맞게 변경)
df = pd.read_csv("titanic.csv")  # Titanic 데이터셋 파일명을 입력하세요.

# 2. 데이터프레임 기본 정보 확인
print("데이터프레임 기본 정보")
df.info()  # 데이터 구조 및 결측치 개수를 확인하는 함수 입력
```

```python
# 3. 숫자형 데이터의 기본 통계 분석
print("\n숫자형 데이터의 요약 통계")
df.describe()  # 숫자형 데이터의 평균, 표준편차, 최소/최대값 등을 확인하는 함수 입력
```

```python
# 4. Pclass(선실 등급)별 요금 비교
# 각 Pclass별 승객 수 확인
print("\n선실 등급(Pclass)별 승객 수:")
df["Pclass"].value_counts()  # Pclass별 승객 수를 계산하는 함수 입력
```

```python
# Pclass별 요금(Fare) 통계 분석
# (1) 1등석 승객의 평균 및 중앙값 요금 확인
first_class_fare_mean = df[df["Pclass"] == 1]["Fare"].mean()  # 평균 요금을 계산하는 함수 입력
first_class_fare_median = df[df["Pclass"] == 1]["Fare"].median()  # 중앙값 요금을 계산하는 함수 입력
print(f"\n1등석 평균 요금: {first_class_fare_mean:.2f}, 중앙값 요금: {first_class_fare_median:.2f}")

# (2) 2등석 승객의 평균 및 중앙값 요금 확인
second_class_fare_mean = df[df["Pclass"] == 2]["Fare"].mean()  # 평균 요금을 계산하는 함수 입력
second_class_fare_median = df[df["Pclass"] == 2]["Fare"].median()  # 중앙값 요금을 계산하는 함수 입력
print(f"\n2등석 평균 요금: {second_class_fare_mean:.2f}, 중앙값 요금: {second_class_fare_median:.2f}")

# (3) 3등석 승객의 평균 및 중앙값 요금 확인
third_class_fare_mean = df[df["Pclass"] == 3]["Fare"].mean()  # 평균 요금을 계산하는 함수 입력
third_class_fare_median = df[df["Pclass"] == 3]["Fare"].median()  # 중앙값 요금을 계산하는 함수 입력
print(f"\n3등석 평균 요금: {third_class_fare_mean:.2f}, 중앙값 요금: {third_class_fare_median:.2f}")
```

```python
# 5. Pclass별 생존자 및 사망자 수 확인
# 생존자(Survived == 1) 및 사망자(Survived == 0)의 각 Pclass별 분포 확인
print("\n선실 등급별 생존자 수:")
print(df[df["Survived"] == 1]["Pclass"].value_counts())  # 생존자의 Pclass 분포를 계산하는 함수 입력

print("\n선실 등급별 사망자 수:")
print(df[df["Survived"] == 0]["Pclass"].value_counts())  # 사망자의 Pclass 분포를 계산하는 함수 입력
```

```python
# 6. 요금(Fare) 결측치 확인 및 처리
print("\n요금(Fare) 컬럼 결측치 개수:")
print(df["Fare"].isna().sum())  # 결측치 개수를 확인하는 함수 입력

# 요금 결측치를 중앙값으로 대체
fare_median = df["Fare"].median()  # 중앙값을 계산하는 함수 입력
df["Fare"].fillna(fare_median, inplace=True)  # 결측치를 중앙값으로 채우는 함수 입력

print("\n결측치 처리 후 요금 컬럼 결측치 개수:")
print(df["Fare"].isna().sum())  # 결측치가 잘 처리되었는지 확인하는 함수 입력
```

```python
# 6. 가중 평균 요금(Fare) 계산
# 가중 평균을 사용하는 이유:
# - 일반 평균(mean)은 모든 데이터를 동일한 가중치로 다룹니다.
# - 하지만 Pclass(선실 등급)에 따라 요금이 다르고, 1등석 승객이 더 비싼 요금을 지불하는 경향이 있음.
# - 따라서 선실 등급별 가중치를 고려하여 가중 평균을 계산하면, 보다 현실적인 요금의 중심 값을 얻을 수 있음.

# Pclass의 역수를 가중치로 사용 (1등석은 가중치가 가장 큼, 3등석은 가장 작음)
df["Pclass_weight"] = 1 / df["Pclass"]  # Pclass의 역수를 가중치로 사용

# 가중 평균 공식:
# Σ(Fare × 가중치) / Σ(가중치)
weighted_fare_avg = (df["Fare"] * df["Pclass_weight"]).sum() / df["Pclass_weight"].sum()
print(f"\n가중 평균 요금(Fare): {weighted_fare_avg:.2f}")
```

```python
# 7. 나이를 가중치로 활용한 새로운 가중 평균 계산
# - 어린 승객(어린이, 청소년)과 성인 승객의 요금 기여도가 다를 가능성이 있음.
# - 따라서 나이를 가중치로 활용하여, 연령대가 다른 승객들이 지불한 요금의 가중 평균을 계산.
df["Age"].fillna(df["Age"].median(), inplace=True)  # 결측치는 중앙값으로 대체

weighted_fare_by_age = (df["Fare"] * df["Age"]).sum() / df["Age"].sum()
print(f"\n나이를 가중치로 고려한 가중 평균 요금(Fare): {weighted_fare_by_age:.2f}")

# 8. 요금(Fare) 결측치 처리
print("\n요금(Fare) 컬럼 결측치 개수:", end=' ')
print(df["Fare"].isna().sum())  # 결측치 개수 확인

# 결측치가 있으면 중앙값으로 대체
fare_median = df["Fare"].median()  # 중앙값 계산
df["Fare"].fillna(fare_median, inplace=True)  # 결측치를 중앙값으로 채움

print("\n결측치 처리 후 요금(Fare) 컬럼 결측치 개수:", end=' ')
print(df["Fare"].isna().sum())  # 결측치가 잘 처리되었는지 확인
```
