```python
# 1. 필요한 라이브러리
import pandas as pd
from scipy.stats import ttest_ind   # (힌트: 두 집단 평균 비교용 t-test 함수)

# 2. 데이터 불러오기
df_ts = pd.read_csv('data/time_series_data.csv')  # (힌트: 상대 경로 사용)
df_ts
df_ts.dtypes

# 3. 날짜 데이터 형식 변환
df_ts["Date"] = pd.to_datetime(df_ts["Date"])  # (힌트: 문자열 → datetime)
df_ts.dtypes

# 4. 날짜 기준 정렬
df_ts = df_ts.sort_values(by='Date')  # (힌트: 날짜 순 정렬)
df_ts

# 5. 기본 통계량
summary_stats = df_ts[['value']].describe()  # (힌트: 'value' 열만 통계 보기)

# 6. 통계량 출력
print("전체 데이터의 기본 통계량")
summary_stats

# 7. 월 정보 추출
df_ts["Month"] = df_ts["Date"].dt.month   # (힌트: 날짜 → 월 추출)
df_ts

# 8. 1월과 12월 데이터 선택
jan_values = df_ts[df_ts["Month"] == 1 ]["value"]  # (힌트: 1월)
dec_values = df_ts[df_ts["Month"] == 12 ]["value"]  # (힌트: 12월)

# 9. 평균 계산
jan_mean = jan_values.mean()
dec_mean = dec_values.mean()

# 10. 평균 출력
print("\n1월과 12월의 평균 비교")
print(f"1월 평균: {jan_mean:.2f}")
print(f"12월 평균: {dec_mean:.2f}")

# 11. 독립표본 t-검정
t_stat, p_value = ttest_ind(jan_values, dec_values, equal_var = False)  # (힌트: 분산 다름 가정)

# 12. t-검정 결과 출력
print("\n독립표본 t-검정 결과")
print(f"t-값: {t_stat:.4f}")
print(f"p-value: {p_value:.4f}")

# 13. 결과 해석
if p_value < 0.05 :   # (힌트: 유의수준 0.05)
    print("1월과 12월의 값이 통계적으로 유의미한 차이가 있습니다!")
else:
    print("1월과 12월의 값에 통계적으로 유의미한 차이가 없습니다.")
    print("연초와 연말의 값이 비슷한 패턴을 보입니다.")
```
