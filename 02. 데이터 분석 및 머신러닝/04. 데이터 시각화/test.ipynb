```python
## hw_4_2
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)  # FutureWarning 제거
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'AppleGothic'   # macOS 기본 한글 폰트
plt.rcParams['axes.unicode_minus'] = False    # 마이너스 기호 깨짐 방지

# 데이터 불러오기
file_path = "data/category_data.csv"
df = pd.read_csv(file_path)  # (힌트: 변수 file_path 사용)

# 데이터 확인
df.head()  # (힌트: 상위 5개 행 확인)
df.info()  # (힌트: 데이터 요약 정보 확인)
df.describe()  # (힌트: 기본 통계량 확인)

# ==========================
# 1. 카테고리별 데이터 분포 시각화 (Bar Chart)
# ==========================

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Category', palette='viridis')  # (힌트: countplot 사용)
plt.title('카테고리별 데이터 분포')
plt.xticks(rotation=45)
plt.xlabel('카테고리')
plt.ylabel('개수')
plt.show()  # (힌트: 그래프 출력)

# ==========================
# 2. 카테고리별 수치 데이터 평균 비교 (Bar Chart)
# ==========================

if 'Value' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.barplot(data=df, x='Category', y='Value', estimator=sum, palette='coolwarm')  # (힌트: barplot, 합계)
    plt.title('카테고리별 값 합계')
    plt.xticks(rotation=45)
    plt.xlabel('카테고리')
    plt.ylabel('합계 값')
    plt.show()

# ==========================
# 3. 카테고리별 분포 (Boxplot)
# ==========================

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Category', y='Value', palette='pastel')  # (힌트: boxplot)
plt.title('카테고리별 값 분포')
plt.xticks(rotation=45)
plt.xlabel('카테고리')
plt.ylabel('값')
plt.show()
```

```python
## hw_4_4
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# 데이터 불러오기
file_path = "data/statistical_data.csv"
df = pd.read_csv(file_path)  # (힌트: 변수 file_path 사용)

# 데이터 확인
df.head()  # (힌트: 상위 5개 행 출력)
df.info()  # (힌트: 데이터 구조 확인)
df.describe()  # (힌트: 기본 통계량 출력)

### 1. 변수 간 관계를 산점도로 시각화 ###
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['변수1'], y=df['변수2'], hue=df['Group'], palette="deep", alpha=0.7)  # (힌트: scatterplot)
plt.title("변수1과 변수2 간 관계 (산점도)")
plt.xlabel("변수1")
plt.ylabel("변수2")
plt.legend(title="Group")  # (힌트: 범례 제목)
plt.show()

### 2. 변수 간 상관 행렬 및 히트맵 ###
correlation_matrix = df[['변수1', '변수2']].corr()  # (힌트: corr 함수)

plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)  # (힌트: heatmap)
plt.title("변수 간 상관 행렬")
plt.show()

### 3. 변수 간 관계를 그룹별로 시각화 (페어플롯) ###
sns.pairplot(df, hue="Group", palette="deep", diag_kind="hist")  # (힌트: pairplot, 히스토그램)
plt.show()

### 4. 그룹별 변수1과 변수2의 분포 비교 (박스플롯) ###
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x="Group", y="변수1", data=df, palette="pastel")  # (힌트: boxplot)
plt.title("그룹별 변수1 분포")

plt.subplot(1, 2, 2)
sns.boxplot(x="Group", y="변수2", data=df, palette="pastel")
plt.title("그룹별 변수2 분포")

plt.tight_layout()  # (힌트: 레이아웃 조정)
plt.show()
```

```python
## ws_4_1
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)  # FutureWarning 제거
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터 로드
file_path = "data/seoul_subway_data.csv"
data = pd.read_csv(file_path)  # (힌트: 변수 file_path 사용)
data.head()  # (힌트: 상위 5개 행 확인)

# 2. 요일별 및 시간대별 이용객 수 집계
weekday_agg = data.groupby('DayOfWeek')['PassengerCount'].sum()  # (힌트: 요일 기준)
time_agg = data.groupby('Time')['PassengerCount'].sum()  # (힌트: 시간 기준)

# 결과 확인
print(weekday_agg)
print(time_agg)

# 3. 요일별 이용객 수 가로 막대 그래프 시각화
plt.figure(figsize=(10, 6))
weekday_agg.plot(kind='barh', color='skyblue')  # (힌트: 가로 막대 barh)
plt.title('Total Passengers by Day of the Week')
plt.xlabel('Passenger Count')
plt.ylabel('Day of the Week')
plt.show()  # (힌트: 그래프 출력)

# 4. 시간대별 이용객 수 꺾은선 그래프 시각화
plt.figure(figsize=(10, 6))
time_agg.plot(kind='line', color='blue', marker='o')  # (힌트: 꺾은선 line)
plt.title('Total Passengers by Time of Day')
plt.xlabel('Time')
plt.ylabel('Passenger Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
## ws_4_2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "data/survey_data.csv"
df = pd.read_csv(file_path)  # (힌트: 변수 file_path 사용)

print("데이터프레임 정보:")
df.info()  # (힌트: 데이터 구조 확인)

print("\n데이터프레임 상위 5개 행:")
df.head()  # (힌트: 상위 5개 행 확인)

df["Gender"] = df["Gender"].map({0: "Female", 1: "Male"})  # (힌트: map 함수 활용)

# 1. 히스토그램 및 KDE 플롯을 활용한 키 분포 분석
plt.figure(figsize=(12, 6))
sns.histplot(df, x="Height", hue="Gender", kde=True, bins=30, alpha=0.5)  # (힌트: histplot)
plt.title("Distribution of Height by Gender (Histogram & KDE)")
plt.xlabel("Height (cm)")
plt.ylabel("Count")
plt.show()

# 2. 박스플롯
plt.figure(figsize=(8, 6))
sns.boxplot(x="Gender", y="Height", data=df)  # (힌트: boxplot)
plt.title("Height Distribution by Gender (Boxplot)")
plt.xlabel("Gender")
plt.ylabel("Height (cm)")
plt.show()

# 3. 성별별 키의 통계값 계산
stats = df.groupby("Gender")["Height"].agg(["mean", "median", "std", "min", "max"])
print("\n성별별 키의 기초 통계:")
stats

# 4. 바이올린 플롯
plt.figure(figsize=(8, 6))
sns.violinplot(x="Gender", y="Height", data=df)  # (힌트: violinplot)
plt.title("Height Distribution by Gender (Violin Plot)")
plt.xlabel("Gender")
plt.ylabel("Height (cm)")
plt.show()
```

```python
## ws_4_3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

gdp_file = "data/gdp_data.csv"
pop_file = "data/population_data.csv"

gdp_df = pd.read_csv(gdp_file)  # (힌트: GDP 데이터 경로)
population_df = pd.read_csv(pop_file)  # (힌트: 인구 데이터 경로)

gdp_df.head()  # (힌트: GDP 상위 5개)
population_df.head()
print(gdp_df.info())  # (힌트: 데이터 구조 확인)
print(population_df.info())

merged_df = pd.merge(gdp_df, population_df, on=['Country', 'Year'])  # (힌트: Country, Year)

merged_df["GDP_per_Capita"] = merged_df["GDP"] / merged_df["Population"]

print("\n병합된 데이터 샘플")
merged_df.head()

# 산점도
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x="Population", y="GDP", hue="Country", alpha=0.7)
plt.xscale("log")  # (힌트: X축 로그 스케일)
plt.yscale("log")
plt.xlabel("Population (Log Scale)")
plt.ylabel("GDP (Log Scale)")
plt.title("GDP vs Population (Log Scale)")
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()

# 선 그래프
plt.figure(figsize=(12, 6))
sns.lineplot(data=merged_df, x="Year", y="GDP_per_Capita", hue="Country", marker="o")  # (힌트: lineplot)
plt.xlabel("Year")
plt.ylabel("GDP per Capita")
plt.title("GDP per Capita Trends Over Time by Country")
plt.grid(True, linestyle="--", linewidth=0.5)
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# 막대 그래프 비교
country_stats = merged_df.groupby("Country").agg({"Population": "mean", "GDP": "mean"}).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=country_stats, x="Country", y="Population", color="lightblue", label="Avg Population")  # (힌트: barplot)
sns.barplot(data=country_stats, x="Country", y="GDP", color="orange", alpha=0.6, label="Avg GDP")
plt.yscale("log")
plt.xlabel("Country")
plt.ylabel("Log Scale Value")
plt.title("Average Population and GDP by Country")
plt.legend()
plt.grid(True, linestyle="--", linewidth=0.5)
plt.show()
```

```python
## ws_4_4
import pandas as pd
import matplotlib.pyplot as plt

file_path = "data/time_series_data.csv"
df = pd.read_csv(file_path)  # (힌트: 변수 file_path 사용)

print(df.info())
df.head()  # (힌트: 상위 5개 행)

window_size = 3
df["Moving_Avg"] = df["Value"].rolling(window=window_size).mean()  # (힌트: rolling 평균)

df["Pct_Change"] = df["Value"].pct_change* 100  # (힌트: pct_change)
df

plt.figure(figsize=(12, 6))
plt.plot(df["Year"], df["Value"], marker='o', linestyle='-', color='b', label="Original Value")
plt.plot(df["Year"], df["Moving_Avg"], marker='s', linestyle='--', color='orange', label=f"{window_size}-Year Moving Avg")
plt.xlabel("Year", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.title("Time Series Data Trend with Moving Average", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

ax2 = plt.gca().twinx()
ax2.bar(df["Year"], df["Pct_Change"], alpha=0.3, color="gray", label="Annual % Change")
ax2.set_ylabel("Percentage Change (%)", fontsize=12)
ax2.legend(loc="upper left")

plt.show()
```
