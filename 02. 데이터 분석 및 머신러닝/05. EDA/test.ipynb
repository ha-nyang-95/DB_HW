```python
## ws_5_2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "data/subway_data.csv"
df = pd.read_csv(file_path)  # [빈칸] 데이터를 불러오는 코드 작성
print("데이터 미리보기")
df.head()  # [빈칸] 데이터의 처음 몇 개의 행 출력
print("\n데이터 정보")
df.info()  # [빈칸] 데이터프레임의 정보를 출력

df = df[['Date', 'Station', 'Passengers']].dropna(axis=0)  # [빈칸]
df['Date'] = pd.to_datetime(df['Date'])  # [빈칸]
df['Weekday'] = df['Date'].dt.day_name()  # [빈칸]

weekday_avg_passengers = df.groupby('Weekday')['Passengers'].mean().reset_index()  # [빈칸]
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_avg_passengers['Weekday'] = pd.Categorical(weekday_avg_passengers['Weekday'], categories=weekday_order, ordered=True)
weekday_avg_passengers = weekday_avg_passengers.sort_values('Weekday')  # [빈칸]

busiest_day = weekday_avg_passengers.loc[weekday_avg_passengers['Passengers'].idxmax()]  # [빈칸]
print("\n가장 평균 승객 수가 많은 요일")
busiest_day  # [빈칸]

plt.figure(figsize=(10, 6))
sns.barplot(data=weekday_avg_passengers, x='Weekday', y='Passengers', palette='viridis')  # [빈칸]
plt.title('요일별 평균 승객 수')
plt.xlabel('요일')  # [빈칸]
plt.ylabel('평균 승객 수')  # [빈칸]
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
## ws_5_3
import pandas as pd
import matplotlib.pyplot as plt

file_path = "data/stock_data_2.csv"
df = pd.read_csv(file_path)  # [빈칸] 데이터를 불러오는 코드 작성
print("데이터 미리보기")
df.head()  # [빈칸] 데이터의 처음 몇 개의 행을 출력
print("\n데이터 정보")
df.info()  # [빈칸] 데이터프레임의 정보를 출력

df = df[['Date', 'Close']].dropna(axis=0)  # [빈칸]
df['Date'] = pd.to_datetime(df['Date'])  # [빈칸]
df = df.sort_values(by='Date')  # [빈칸]

df['MA5'] = df['Close'].rolling(window=5).mean()  # [빈칸]
df['MA20'] = df['Close'].rolling(window=20).mean()  # [빈칸]
df['MA60'] = df['Close'].rolling(window=60).mean()  # [빈칸]

df['Crossover'] = df['MA5'] - df['MA20']  # [빈칸]

golden_cross = df[(df['Crossover'] > 0) & (df['Crossover'].shift(1) < 0)]  # [빈칸]
death_cross = df[(df['Crossover'] < 0) & (df['Crossover'].shift(1) > 0)]  # [빈칸]

print("골든크로스")
golden_cross
print("데드크로스")
death_cross

plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], label='종가', color='black', linewidth=1, alpha=0.7)  # [빈칸]
plt.plot(df['Date'], df['MA5'], label='5일 이동 평균선', color='blue', linewidth=1)  # [빈칸]
plt.plot(df['Date'], df['MA20'], label='20일 이동 평균선', color='orange', linewidth=1)  # [빈칸]
plt.plot(df['Date'], df['MA60'], label='60일 이동 평균선', color='green', linewidth=1)  # [빈칸]
plt.scatter(golden_cross['Date'], golden_cross['Close'], color='red', label='골든크로스', marker='^', s=100)  # [빈칸]
plt.scatter(death_cross['Date'], death_cross['Close'], color='purple', label='데드크로스', marker='v', s=100)  # [빈칸]
plt.title('주식 데이터의 이동 평균선 분석 및 특정 구간 강조')
plt.xlabel('Date')  # [빈칸]
plt.ylabel('Close')  # [빈칸]
plt.legend()
plt.grid(True)
plt.show()  # [빈칸]
```

```python
## ws_5_4
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'

file_path = "data/store_sales.csv"
df = pd.read_csv(file_path)  # [빈칸] 데이터를 불러오는 코드 작성
print("데이터 미리보기")
df.head()  # [빈칸]
print("\n데이터 정보")
df.info()  # [빈칸]

df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m")
df["YearMonth"] = df["Date"].dt.to_period("M")

monthly_sales = df.groupby("YearMonth")["Revenue"].sum().reset_index()
monthly_sales["YearMonth"] = monthly_sales["YearMonth"].astype("str")

product_sales = df.groupby("Product")["Quantity"].sum().reset_index()
top_products = product_sales.sort_values(by="Quantity", ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x="YearMonth", y="Revenue", marker="o")
plt.xticks(rotation=45)
plt.title("월별 매출 변화")
plt.xlabel("_____")  # [빈칸]
plt.ylabel("_____")  # [빈칸]
plt.grid()
plt._____()  # [빈칸]

plt.figure(figsize=(12, 6))
sns.barplot(data=top_products, x="Product", y="Quantity")
plt.title("가장 많이 팔린 제품 상위 10개")
plt.xlabel("_____")  # [빈칸]
plt.ylabel("_____")  # [빈칸]
plt.xticks(rotation=45)
plt._____()  # [빈칸]
```
