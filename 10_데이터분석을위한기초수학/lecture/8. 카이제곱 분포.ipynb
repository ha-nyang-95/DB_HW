```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. 카이제곱 분포를 따르는 난수 생성
data_chi_square = np.random.chisquare(df=1, size=1000)

"""
np.random.chisquare(df=1, size=1000)
- 자유도(df=1)를 가지는 카이제곱 분포에서 1000개의 난수를 생성함.
- 카이제곱 분포는 비대칭(right-skewed) 형태이며, 자유도(df)에 따라 모양이 달라짐.
- df가 작을수록 오른쪽으로 긴 꼬리를 가지는 형태이고, df가 커질수록 정규 분포에 가까워짐.
"""

# 2. 그래프 크기 설정
plt.figure(figsize=(12, 8)) # 가로 12, 세로 8 크기의 그래프 설정

# 3. 카이제곱 분포 데이터를 히스토그램으로 시각화
sns.histplot(data_chi_square, kde=True, color='gold')

"""
- sns.histplot(): 데이터의 히스토그램을 그리는 함수.
- kde=True: 부드러운 확률 밀도 곡선(Kernel Density Estimation, KDE)을 추가하여 분포 형태를 더 잘 보여줌.
- color='gold': 그래프 색상을 금색(gold)으로 설정.
"""

#4. 그래프 제목 및 축 라벨 설정
plt.title('Chi-Square Distribution (df=1)') # 그래프 제목
plt.xlabel('Value') # X축 라벨 (데이터 값)
plt.ylabel('Frequency') # Y축 라벨(빈도수)

#5. 그래프 출력
plt.show()
```
