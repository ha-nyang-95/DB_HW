```python
import numpy as np
import pandas as pd
from scipy.stats import norm

# 1. ì •ê·œë¶„í¬ ë°ì´í„° ìƒì„±
# - í‰ê· (mu): 50
# - í‘œì¤€í¸ì°¨(sigma): 10
# - ë°ì´í„° ê°œìˆ˜(size): 1000ê°œ
# - numpyì˜ random.normal(í‰ê· , í‘œì¤€í¸ì°¨, ë°ì´í„° ê°œìˆ˜) 
#   ì •ê·œë¶„í¬ í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ë°ì´í„°ë¥¼ ìƒì„±í•˜ì„¸ìš”.

mu, sigma, size = 50, 10, 1000
data = np.random.normal(mu, sigma, size)  # ë¹ˆì¹¸ì„ ì±„ì›Œ ì •ê·œë¶„í¬ ë°ì´í„° ìƒì„±
print(data)
```

```python
# 2. ë°ì´í„°ë¥¼ DataFrameìœ¼ë¡œ ë³€í™˜
# - pandasì˜ DataFrame()ì„ ì‚¬ìš©í•˜ì—¬ ë°ì´í„°ë¥¼ ì €ì¥í•˜ì„¸ìš”.
df = pd.DataFrame({"value": data})  # ë¹ˆì¹¸ì„ ì±„ì›Œ ë°ì´í„°í”„ë ˆì„ ìƒì„±
print(df)
```

```python
# 3. ì „ì²´ ë°ì´í„°ì˜ ê¸°ë³¸ í†µê³„ëŸ‰ í™•ì¸
# - describe() í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ í‰ê· , í‘œì¤€í¸ì°¨, ìµœì†Œê°’, ìµœëŒ€ê°’ ë“±ì˜ í†µê³„ë¥¼ í™•ì¸í•˜ì„¸ìš”.
summary_stats = df["value"].describe()  # ë¹ˆì¹¸ì„ ì±„ì›Œ describe() í•¨ìˆ˜ í˜¸ì¶œ
print(summary_stats)
```

```python
# 4. ì •ê·œë¶„í¬ì˜ ì´ë¡ ì  ê¸°ëŒ€ê°’ê³¼ ë¹„êµ
# - ì •ê·œë¶„í¬ëŠ” í‰ê· (mu)ê³¼ í‘œì¤€í¸ì°¨(sigma)ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë°ì´í„°ê°€ ë¶„í¬í•¨
# - 68-95-99.7 ë²•ì¹™ì„ ì ìš©í•˜ì—¬ ì‹¤ì œ ë°ì´í„°ê°€ ì •ê·œë¶„í¬ì™€ ì–¼ë§ˆë‚˜ ìœ ì‚¬í•œì§€ í™•ì¸

# í‰ê· ê³¼ í‘œì¤€í¸ì°¨ ê³„ì‚°
# - ë°ì´í„°ì˜ í‰ê· ê³¼ í‘œì¤€í¸ì°¨ë¥¼ ê³„ì‚°í•˜ì„¸ìš”.
mean_value = df["value"].mean()  # ë¹ˆì¹¸ì„ ì±„ì›Œ í‰ê·  ê³„ì‚°
std_value = df["value"].std()  # ë¹ˆì¹¸ì„ ì±„ì›Œ í‘œì¤€í¸ì°¨ ê³„ì‚°

# 68-95-99.7 ë²•ì¹™ ì ìš©
# - ì •ê·œë¶„í¬ì—ì„œëŠ” í‰ê· ì„ ê¸°ì¤€ìœ¼ë¡œ ë°ì´í„°ê°€ íŠ¹ì • ë²”ìœ„ ë‚´ì— í¬í•¨ë˜ëŠ” ë¹„ìœ¨ì´ ì •í•´ì ¸ ìˆìŒ
# - í‰ê· (mu) Â±1 í‘œì¤€í¸ì°¨(sigma) ë‚´ì— ì•½ 68%ì˜ ë°ì´í„°ê°€ ì¡´ì¬í•´ì•¼ í•¨
# - í‰ê·  Â±2 í‘œì¤€í¸ì°¨ ë‚´ì— ì•½ 95%ì˜ ë°ì´í„°ê°€ ì¡´ì¬í•´ì•¼ í•¨
# - í‰ê·  Â±3 í‘œì¤€í¸ì°¨ ë‚´ì— ì•½ 99.7%ì˜ ë°ì´í„°ê°€ ì¡´ì¬í•´ì•¼ í•¨

within_1_sigma = np.sum((df["value"] >= mu - sigma) & (df["value"] <= mu + sigma)) / size * 100  # Â±1 í‘œì¤€í¸ì°¨
within_2_sigma = np.sum((df["value"] >= mu - 2*sigma) & (df["value"] <= mu + 2*sigma)) / size * 100  # Â±2 í‘œì¤€í¸ì°¨
within_3_sigma = np.sum((df["value"] >= mu - 3*sigma) & (df["value"] <= mu + 3*sigma)) / size * 100  # Â±3 í‘œì¤€í¸ì°¨

print(within_1_sigma)
print(within_2_sigma)
print(within_3_sigma)
```

```python
# 5. ì •ê·œë¶„í¬ ê²€ì • ìˆ˜í–‰
# - í‰ê· ê³¼ í‘œì¤€í¸ì°¨ê°€ ì •ê·œë¶„í¬ì˜ ì´ë¡ ì ì¸ ê°’ê³¼ ì–¼ë§ˆë‚˜ ìœ ì‚¬í•œì§€ í™•ì¸
# - íŠ¹ì • ê°’(ì˜ˆ: í‰ê· )ì—ì„œì˜ í™•ë¥ ë°€ë„í•¨ìˆ˜(PDF) ê°’ì„ ê³„ì‚°í•˜ì—¬ ì •ê·œì„±ì„ ê²€í† 
# - pdf(í™•ë¥ ë³€ìˆ˜ê°’, í‰ê· , í‘œì¤€í¸ì°¨) => í™•ë¥ ë³€ìˆ˜ê°’ì— ëŒ€í•œ ì •ê·œí™•ë¥ ë°€ë„ê°’

sample_value = mu  # ë¹ˆì¹¸ì„ ì±„ì›Œ íŠ¹ì • ê°’ì„ ì„¤ì •
pdf_value = norm.pdf(sample_value, loc=mu, scale=sigma)  # ë¹ˆì¹¸ì„ ì±„ì›Œ PDF ê³„ì‚°
print(pdf_value)
```

```python
# 6. ê²°ê³¼ ì¶œë ¥ ë° í•´ì„
# - ìƒì„±ëœ ë°ì´í„°ê°€ í‰ê·  50, í‘œì¤€í¸ì°¨ 10ì„ ë”°ë¥´ëŠ” ì •ê·œë¶„í¬ì— ê°€ê¹ë‹¤ëŠ” ê²ƒì„ í™•ì¸
# - 68-95-99.7 ë²•ì¹™ì„ ì ìš©í•œ ê²°ê³¼, ì‹¤ì œ ë°ì´í„°ê°€ ì´ë¡ ì  ì •ê·œë¶„í¬ì™€ ìœ ì‚¬í•œ ë¶„í¬ë¥¼ ë³´ì„
# - ìƒ˜í”Œ ë°ì´í„°ì˜ í‰ê· ê³¼ í‘œì¤€í¸ì°¨ê°€ ì •ê·œë¶„í¬ì˜ ì´ë¡ ì  ê°’ê³¼ ê±°ì˜ ë™ì¼í•¨
# - ì¶”ê°€ì ì¸ ì •ê·œì„± ê²€ì •(ì˜ˆ: QQ-Plot, ìƒ¤í”¼ë¡œ-ìœŒí¬ ê²€ì • ë“±)ì„ ìˆ˜í–‰í•˜ë©´ ë³´ë‹¤ ì •í™•í•œ ë¶„ì„ì´ ê°€ëŠ¥í•¨

print("ì „ì²´ ë°ì´í„°ì˜ ê¸°ë³¸ í†µê³„ëŸ‰:")
print(summary_stats)

print("\n68-95-99.7 ë²•ì¹™ ì ìš© ê²°ê³¼:")
print(f"68% ë²”ìœ„ ë‚´ ë°ì´í„° ë¹„ìœ¨: {within_1_sigma:.2f}%")
print(f"95% ë²”ìœ„ ë‚´ ë°ì´í„° ë¹„ìœ¨: {within_2_sigma:.2f}%")
print(f"99.7% ë²”ìœ„ ë‚´ ë°ì´í„° ë¹„ìœ¨: {within_3_sigma:.2f}%")

print("\nì •ê·œë¶„í¬ ê²€ì • ê²°ê³¼:")
print(f"ìƒ˜í”Œ í‰ê· : {mean_value:.2f}, ìƒ˜í”Œ í‘œì¤€í¸ì°¨: {std_value:.2f}")
print(f"ì´ë¡ ì  ì •ê·œë¶„í¬ì˜ í‰ê· : {mu}, í‘œì¤€í¸ì°¨: {sigma}")
print(f"í‰ê·  ê°’({sample_value})ì—ì„œì˜ ì´ë¡ ì  í™•ë¥ ë°€ë„í•¨ìˆ˜(PDF) ê°’: {pdf_value:.6f}")
```

## ws_3_3

```python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'

# 1. CSV íŒŒì¼ì„ ë¶ˆëŸ¬ì™€ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ì €ì¥
# - pandasì˜ read_csv() í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¤ì„¸ìš”.
file_path = "data_analysis1_ws_3_3\\data\\statistics_data.csv"
df = pd.read_csv(file_path)
print(df)
```

```python
# 2. ë°ì´í„° ì •ë³´ í™•ì¸ (ì»¬ëŸ¼ëª…, ë°ì´í„° íƒ€ì…, ê²°ì¸¡ì¹˜ ì—¬ë¶€ ë“±)
# - info() í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ë°ì´í„°ì˜ êµ¬ì¡°ë¥¼ í™•ì¸í•˜ì„¸ìš”.
print(df.info())
```

```python
# 3. ë°•ìŠ¤í”Œë¡¯ì„ ìœ„í•œ í™•ì¥ëœ í†µê³„ ë¶„ì„ ìˆ˜í–‰
# - ë°ì´í„°ì˜ ì´ìƒì¹˜ íƒìƒ‰ê³¼ í•¨ê»˜, ê° ì»¬ëŸ¼ì˜ ë³€ë™ì„± ë° ì¤‘ì‹¬ ê²½í–¥ì„±ì„ í™•ì¸í•©ë‹ˆë‹¤.

# ë¶„ì„í•  ì»¬ëŸ¼ ë¦¬ìŠ¤íŠ¸ ìƒì„±
columns_to_analyze = df.columns
print(columns_to_analyze)
```

```python
# 4. ê° ì»¬ëŸ¼ë³„ ì‚¬ë¶„ìœ„ìˆ˜(Quartiles), IQR ë° ì´ìƒì¹˜ ë¶„ì„

# ê²°ê³¼ ì €ì¥ìš© ë”•ì…”ë„ˆë¦¬ ìƒì„±
boxplot_stats = {}

for column in columns_to_analyze:
    # Q1 (ì œ1ì‚¬ë¶„ìœ„ìˆ˜, 25%)
    Q1 = df[column].quantile(.25)
    
    # Q2 (ì¤‘ì•™ê°’, 50%)
    Q2 = df[column].median()
    
    # Q3 (ì œ3ì‚¬ë¶„ìœ„ìˆ˜, 75%)
    Q3 = df[column].quantile(.75)
    
    # IQR (ì‚¬ë¶„ìœ„ ë²”ìœ„, Interquartile Range)
    IQR = Q3 - Q1
    
    # ì´ìƒì¹˜ íƒìƒ‰ ê¸°ì¤€ (1.5 * IQR ë° 3.0 * IQR ê¸°ì¤€ ì ìš©)
    lower_bound_1_5 = Q1 - 1.5 * IQR  # ì¼ë°˜ ì´ìƒì¹˜ í•˜í•œ
    upper_bound_1_5 = Q3 + 1.5 * IQR  # ì¼ë°˜ ì´ìƒì¹˜ ìƒí•œ
    
    lower_bound_3_0 = Q1 - 3.0 * IQR  # ê·¹ë‹¨ ì´ìƒì¹˜ í•˜í•œ
    upper_bound_3_0 = Q3 + 3.0 * IQR  # ê·¹ë‹¨ ì´ìƒì¹˜ ìƒí•œ
    
    # ì¼ë°˜ ì´ìƒì¹˜ ê°œìˆ˜ (1.5 * IQR ê¸°ì¤€)
    mild_outliers = df[(df[column] < lower_bound_1_5) | (df[column] > upper_bound_1_5)]
    
    # ê·¹ë‹¨ì  ì´ìƒì¹˜ ê°œìˆ˜ (3.0 * IQR ê¸°ì¤€)
    extreme_outliers = df[(df[column] < lower_bound_3_0) | (df[column] > upper_bound_3_0)]
    
    # ì»¬ëŸ¼ë³„ ë¶„ì„ ê²°ê³¼ ì €ì¥
    boxplot_stats[column] = {
        "Q1": Q1,
        "Q2 (Median)": Q2,
        "Q3": Q3,
        "IQR": IQR,
        "Lower Bound (1.5 * IQR)": lower_bound_1_5,
        "Upper Bound (1.5 * IQR)": upper_bound_1_5,
        "Lower Bound (3.0 * IQR)": lower_bound_3_0,
        "Upper Bound (3.0 * IQR)": upper_bound_3_0,
        "Mild Outliers Count (1.5 * IQR)": len(mild_outliers),
        "Extreme Outliers Count (3.0 * IQR)": len(extreme_outliers),
        "Variance": np.var(df[column]),
        "Standard Deviation": np.std(df[column]),
        "Mean Absolute Deviation": np.mean(np.abs(df[column] - df[column].median()))
    }
```

```python
# 5. ë¶„ì„ ê²°ê³¼ ì¶œë ¥
for column, stats in boxplot_stats.items():
    print(f"\n{column} ì»¬ëŸ¼ì˜ ë°•ìŠ¤í”Œë¡¯ í†µê³„ ë° ì´ìƒì¹˜ ë¶„ì„:")
    print(f"  - ì œ1ì‚¬ë¶„ìœ„ìˆ˜(Q1, 25%): {stats['Q1']:.2f}")
    print(f"  - ì¤‘ì•™ê°’(Q2, 50%): {stats['Q2 (Median)']:.2f}")
    print(f"  - ì œ3ì‚¬ë¶„ìœ„ìˆ˜(Q3, 75%): {stats['Q3']:.2f}")
    print(f"  - IQR (Q3 - Q1): {stats['IQR']:.2f}")
    print(f"  - ì¼ë°˜ ì´ìƒì¹˜ í•˜í•œ (1.5 * IQR): {stats['Lower Bound (1.5 * IQR)']:.2f}")
    print(f"  - ì¼ë°˜ ì´ìƒì¹˜ ìƒí•œ (1.5 * IQR): {stats['Upper Bound (1.5 * IQR)']:.2f}")
    print(f"  - ê·¹ë‹¨ ì´ìƒì¹˜ í•˜í•œ (3.0 * IQR): {stats['Lower Bound (3.0 * IQR)']:.2f}")
    print(f"  - ê·¹ë‹¨ ì´ìƒì¹˜ ìƒí•œ (3.0 * IQR): {stats['Upper Bound (3.0 * IQR)']:.2f}")
    print(f"  - ì¼ë°˜ ì´ìƒì¹˜ ê°œìˆ˜ (1.5 * IQR ê¸°ì¤€): {stats['Mild Outliers Count (1.5 * IQR)']}ê°œ")
    print(f"  - ê·¹ë‹¨ ì´ìƒì¹˜ ê°œìˆ˜ (3.0 * IQR ê¸°ì¤€): {stats['Extreme Outliers Count (3.0 * IQR)']}ê°œ")
    print(f"  - ë¶„ì‚° (Variance): {stats['Variance']:.2f}")
    print(f"  - í‘œì¤€í¸ì°¨ (Standard Deviation): {stats['Standard Deviation']:.2f}")
    print(f"  - í‰ê·  ì ˆëŒ€ í¸ì°¨ (Mean Absolute Deviation): {stats['Mean Absolute Deviation']:.2f}")
```

## ws_3_5

```python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'

# 1. CSV íŒŒì¼ì„ ë¶ˆëŸ¬ì™€ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ì €ì¥
file_path = "data_analysis1_ws_3_5\\data\\eda_data.csv"
df = pd.read_csv(file_path)
print(df)
print()
```

```python
print('## Old value ##')
print()
print(df.dtypes)
print()
print(df.isnull().sum())
print()

# ë‚ ì§œ ë°ì´í„°ë¥¼ datetime íƒ€ì…ìœ¼ë¡œ ë³€í™˜í•˜ì—¬ ì‹œê°„ ë°ì´í„°ë¡œ í™œìš© ê°€ëŠ¥í•˜ê²Œ í•¨
df['Date'] = pd.to_datetime(df['Date'])  # ë¹ˆì¹¸ ì±„ìš°ê¸°

# ê²°ì¸¡ê°’ í™•ì¸ ë° ì²˜ë¦¬ (ê²°ì¸¡ê°’ì´ ìˆëŠ” ê²½ìš° í‰ê· ê°’ìœ¼ë¡œ ëŒ€ì²´)
# í‰ê· ê°’ìœ¼ë¡œ ê²°ì¸¡ì¹˜ë¥¼ ëŒ€ì²´í•˜ë©´ ë°ì´í„°ì˜ ì™œê³¡ì„ ìµœì†Œí™”í•  ìˆ˜ ìˆìŒ
df = df.fillna(df.mean())  # ë¹ˆì¹¸ ì±„ìš°ê¸°

print('## New Value ##')
print()
print(df.dtypes)
print()
print(df.isnull().sum())
```

```python
# ì´ìƒì¹˜ íƒì§€ ë° ì œê±° (ì‚¬ë¶„ìœ„ìˆ˜ ë²”ìœ„ë¥¼ ì´ìš©í•œ IQR ë°©ë²• ì‚¬ìš©)
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(.25)  # 1ì‚¬ë¶„ìœ„ ê°’ (Q1) ë¹ˆì¹¸ ì±„ìš°ê¸°
        Q3 = df[col].quantile(.75)  # 3ì‚¬ë¶„ìœ„ ê°’ (Q3) ë¹ˆì¹¸ ì±„ìš°ê¸°
        IQR = Q3 - Q1  # ì‚¬ë¶„ìœ„ ë²”ìœ„ (Interquartile Range, IQR)
        lower_bound = Q1 - 1.5 * IQR  # ì´ìƒì¹˜ í•˜í•œê°’
        upper_bound = Q3 + 1.5 * IQR  # ì´ìƒì¹˜ ìƒí•œê°’
        # ì§€ì •ëœ ë²”ìœ„ë¥¼ ë²—ì–´ë‚œ ì´ìƒì¹˜ë¥¼ ì œê±°
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]  # ë¹ˆì¹¸ ì±„ìš°ê¸°
    return df

# ìˆ˜ì¹˜í˜• ë³€ìˆ˜ì— ëŒ€í•´ ì´ìƒì¹˜ ì œê±° ìˆ˜í–‰
df = remove_outliers(df, ['column_1', 'column_2', 'column_3', 'column_4'])  # ë¹ˆì¹¸ ì±„ìš°ê¸°

# ë°ì´í„°ì˜ ê¸°ì´ˆ í†µê³„ëŸ‰ í™•ì¸
# count(ê°œìˆ˜), mean(í‰ê· ), std(í‘œì¤€í¸ì°¨), min(ìµœì†Ÿê°’), max(ìµœëŒ“ê°’) ë“± ì œê³µ
descriptive_stats = df.dtypes  # ë¹ˆì¹¸ ì±„ìš°ê¸°
print("ê¸°ì´ˆ í†µê³„ëŸ‰:")
print(descriptive_stats)

# ë³€ìˆ˜ ê°„ì˜ ìƒê´€ ê³„ìˆ˜ ê³„ì‚°
# ìƒê´€ ê³„ìˆ˜ëŠ” -1ì—ì„œ 1 ì‚¬ì´ì˜ ê°’ì„ ê°€ì§€ë©°, 
# 1ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ê°•í•œ ì–‘ì˜ ìƒê´€ê´€ê³„, 
# -1ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ê°•í•œ ìŒì˜ ìƒê´€ê´€ê³„ë¥¼ ì˜ë¯¸
correlation_matrix = df.corr()  # ë¹ˆì¹¸ ì±„ìš°ê¸°
print("\nìƒê´€ ê³„ìˆ˜ í–‰ë ¬:")
print(correlation_matrix)
```

```python
# ìƒê´€ ê³„ìˆ˜ í–‰ë ¬ì„ ì‹œê°í™”í•˜ì—¬ ë³€ìˆ˜ ê°„ì˜ ê´€ê³„ë¥¼ ì‰½ê²Œ íŒŒì•…
plt.figure(figsize=(8, 6))  # íˆíŠ¸ë§µ í¬ê¸° ì„¤ì •
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("ë³€ìˆ˜ ê°„ì˜ ìƒê´€ ê´€ê³„ íˆíŠ¸ë§µ")
plt.show()
```

```python
# ë³€ìˆ˜ ê°„ì˜ ê´€ê³„ë¥¼ ì‹œê°í™”í•˜ê¸° ìœ„í•´ í˜ì–´í”Œë¡¯(pairplot) ìƒì„±
# ============= í˜ì–´í”Œë¡¯ ============= 
# ì—¬ëŸ¬ ë³€ìˆ˜ ê°„ì˜ ê´€ê³„ë¥¼ í•œ ë²ˆì— ì‹œê°ì ìœ¼ë¡œ ë¶„ì„í•˜ëŠ” ë° ìœ ìš©í•¨
# ê° ë³€ìˆ˜ ê°„ì˜ ì‚°ì ë„ë¥¼ ê·¸ë ¤ ê´€ê³„ë¥¼ íŒŒì•…í•  ìˆ˜ ìˆìœ¼ë©°, ëŒ€ê°ì„ ì—ëŠ” ê° ë³€ìˆ˜ì˜ ë¶„í¬ë¥¼ í‘œì‹œí•¨
# kind='reg'ë¥¼ ì‚¬ìš©í•˜ì—¬ ì‚°ì ë„ì— ì„ í˜• íšŒê·€ì„ ì„ ì¶”ê°€í•˜ì—¬ ê´€ê³„ì˜ ë°©í–¥ì„ ì‹œê°ì ìœ¼ë¡œ ê°•ì¡°
# diag_kind='kde'ë¥¼ ì‚¬ìš©í•˜ì—¬ ëŒ€ê°ì„ ì˜ ë¶„í¬ë¥¼ ì»¤ë„ ë°€ë„ ê·¸ë˜í”„ë¡œ í‘œì‹œí•˜ì—¬ ë³€ìˆ˜ì˜ ë¶„í¬ í˜•íƒœë¥¼ ì§ê´€ì ìœ¼ë¡œ í™•ì¸
sns.pairplot(df.drop(columns=['Date']), kind='reg', diag_kind='kde')
plt.suptitle("ë³€ìˆ˜ ê°„ì˜ ê´€ê³„ ì‹œê°í™” (Pairplot with Regression)", y=1.02)  # ì œëª© ì„¤ì •
plt.show()
```
