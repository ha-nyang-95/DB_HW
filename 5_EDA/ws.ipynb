```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 데이터 불러오기 및 기본 탐색
# 주어진 지하철 데이터 파일을 불러온다.
file_path = "data/subway_data.csv"  # 실제 데이터 파일 경로
df = pd.read_csv(file_path)  # [빈칸] 데이터를 불러오는 코드 작성

# 데이터의 첫 5행 출력 (EDA의 첫 단계)
print("데이터 미리보기")
df.head()  # [빈칸] 데이터의 처음 몇 개의 행을 출력하는 코드 작성
```

```python
# 데이터의 기본 정보 확인 (컬럼명, 데이터 타입, 결측값 확인)
print("\n데이터 정보")
df.info()  # [빈칸] 데이터프레임의 정보를 출력하는 코드 작성
```

```python
# 2. 필요한 컬럼 선택 및 데이터 변환
# 'Date' (날짜), 'Station' (지하철역), 'Passengers' (승객 수) 컬럼이 존재한다고 가정하고, 필요한 컬럼만 선택
df = df[['Date', 'Station', 'Passengers']].dropna(axis=0)  
# [빈칸] 필요한 컬럼만 선택하고 결측값이 있는 행을 제거하는 코드 작성

# 'Date' 컬럼을 날짜 형식으로 변환하여 요일 정보 추출
df['Date'] = pd.to_datetime(df['Date'])  # [빈칸] 'Date' 컬럼을 datetime 형식으로 변환
df['Weekday'] = df['Date'].dt.day_name()  # [빈칸] 날짜 데이터에서 요일(월요일~일요일)을 추출하는 코드 작성

# 3. 요일별 평균 승객 수 분석
# 요일별 평균 승객 수를 계산하여 새로운 데이터프레임 생성
weekday_avg_passengers = df.groupby('Weekday')['Passengers'].mean().reset_index()  
# [빈칸] 'Weekday'를 기준으로 'Passengers'의 평균을 계산하는 코드 작성
weekday_avg_passengers
```

```python
# 요일별 데이터를 정렬 (월요일~일요일 순서)
# sort_values() 함수는 기본적으로 열의 데이터를 오름차순(또는 내림차순)으로 정렬합니다. 
# 그러나 'Weekday' 컬럼의 경우 단순 문자열 데이터라면 알파벳 순서로 정렬됩니다. 
# 예를 들어, 알파벳 순으로 정렬하면 
# "Friday", "Monday", "Saturday", "Sunday", "Thursday", "Tuesday", "Wednesday" 순서로 나올 수 있습니다.

# 따라서, 요일을 "월요일~일요일" 순서로 정렬하려면 
# 먼저 'Weekday' 컬럼을 pd.Categorical 타입으로 변환하여 원하는 순서를 명시해야 합니다. 
# 이렇게 하면 sort_values()가 이 사용자 정의 순서를 따르게 됩니다.

weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_avg_passengers['Weekday'] = pd.Categorical(weekday_avg_passengers['Weekday'], categories=weekday_order, ordered=True)
weekday_avg_passengers = weekday_avg_passengers.sort_values('Weekday')  # [빈칸] 요일 기준으로 정렬하는 코드 작성
weekday_avg_passengers
```

```python
# 4. 가장 승객 수가 많은 요일 찾기
busiest_day = weekday_avg_passengers.loc[weekday_avg_passengers['Passengers'].idxmax()]  
# [빈칸] 가장 평균 승객 수가 많은 요일을 찾는 코드 작성

print("\n가장 평균 승객 수가 많은 요일")
busiest_day  # [빈칸] 가장 승객 수가 많은 요일 데이터를 출력하는 코드 작성
```

```python
# 5. 데이터 시각화 (요일별 평균 승객 수 비교)
plt.figure(figsize=(10, 6))  
sns.barplot(data=weekday_avg_passengers, x='Weekday', y='Passengers', palette='viridis')  
# [빈칸] Seaborn의 barplot을 사용하여 요일별 평균 승객 수를 시각화하는 코드 작성

# 그래프 제목 및 라벨 설정
plt.title('요일별 평균 승객 수')
plt.xlabel('요일')  # [빈칸] X축 라벨 설정
plt.ylabel('평균 승객 수')  # [빈칸] Y축 라벨 설정
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 6. 결과 해석
# 1. 특정 요일별로 평균 승객 수를 분석하여 가장 붐비는 요일을 찾았다.
# 2. 전체적으로 주중(월요일~금요일)이 주말보다 승객 수가 많을 가능성이 높음.
# 3. 가장 승객 수가 많은 요일이 금요일이라면, 이는 출퇴근 및 여가 활동이 겹치는 영향을 받을 가능성이 있음.
# 4. 주말의 경우(토요일, 일요일)는 상대적으로 승객 수가 감소할 수 있으며, 특정 관광지가 포함된 역은 예외일 수 있음.
# 5. 가장 붐비는 요일의 역별 승객 수를 추가적으로 분석하여, 특정 역이 평균을 끌어올리는지 확인할 필요가 있음.

# 그래프 출력
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터 불러오기 및 기본 탐색
# 주어진 주식 데이터 파일을 불러온다.
file_path = "data/stock_data_2.csv"  # 실제 데이터 파일 경로
df = pd.read_csv(file_path)  # [빈칸] 데이터를 불러오는 코드 작성

# 데이터의 첫 5행 출력 (EDA의 첫 단계)
print("데이터 미리보기")
df.head()  # [빈칸] 데이터의 처음 몇 개의 행을 출력하는 코드 작성
```

```python
# 데이터의 기본 정보 확인 (컬럼명, 데이터 타입, 결측값 확인)
print("\n데이터 정보")
df.info()  # [빈칸] 데이터프레임의 정보를 출력하는 코드 작성
```

```python
# 2. 필요한 컬럼 선택 및 데이터 변환
# 'Date' (날짜), 'Close' (종가) 컬럼이 존재한다고 가정하고, 필요한 컬럼만 선택
df = df[['Date', 'Close']].dropna(axis=0)  
# [빈칸] 필요한 컬럼만 선택하고 결측값이 있는 행을 제거하는 코드 작성

# 'Date' 컬럼을 날짜 형식으로 변환
df['Date'] = pd.to_datetime(df['Date'])  # [빈칸] 'Date' 컬럼을 datetime 형식으로 변환

# 날짜 기준으로 정렬
df = df.sort_values(by='Date')  # [빈칸] 날짜 기준으로 정렬하는 코드 작성

# 3. 이동 평균선 계산
# 단기 이동 평균선 (5일)
df['MA5'] = df['Close'].rolling(window=5).mean()  # [빈칸] 5일 이동 평균선 계산

# 중기 이동 평균선 (20일)
df['MA20'] = df['Close'].rolling(window=20).mean()  # [빈칸] 20일 이동 평균선 계산

# 장기 이동 평균선 (60일)
df['MA60'] = df['Close'].rolling(window=60).mean()  # [빈칸] 60일 이동 평균선 계산

df.tail()
```

```python
# 4. 특정 구간 강조 (이동 평균선이 교차하는 구간)
# 이동 평균선이 교차하는 특정 구간을 찾기 위해, MA5와 MA20의 차이를 계산
df['Crossover'] = df['MA5'] - df['MA20']  # [빈칸] 이동 평균선 차이를 계산하는 코드 작성
df.tail()
```

```python
# 골든크로스 (Golden Cross) 찾기
# - 골든크로스란? 단기 이동 평균선(MA5)이 중기 이동 평균선(MA20)을 
#   "상향 돌파"하는 지점을 의미한다.
# - 즉, 오늘의 MA5가 MA20보다 크고, 이전 날의 MA5는 MA20보다 작았던 경우 발생한다.
golden_cross = df[(df['Crossover'] > 0) & (df['Crossover'].shift(1) < 0)]  
# shift(1)은 이전 날의 데이터를 가져온다.
# 따라서, 현재 행의 Crossover 값이 양수이고, 
# 이전 날의 Crossover 값이 음수라면 교차가 발생한 것이다.

# 데드크로스 (Death Cross) 찾기
# - 데드크로스란? 단기 이동 평균선(MA5)이 중기 이동 평균선(MA20)을 
# "하향 돌파"하는 지점을 의미한다.
# - 즉, 오늘의 MA5가 MA20보다 작고, 이전 날의 MA5는 MA20보다 컸던 경우 발생한다.
death_cross = df[(df['Crossover'] < 0) & (df['Crossover'].shift(1) > 0)]  
# shift(1)을 통해 이전 날의 Crossover 값을 확인하여, 교차점이 발생하는 지점을 식별한다.

print(f'골든크로스 (Golden Cross)')
golden_cross
```

```python
print(f'데드크로스 (Death Cross)')
death_cross
```

```python
# 5. 데이터 시각화 (이동 평균선 및 특정 구간 강조)
plt.figure(figsize=(12, 6))

# 종가 그래프
plt.plot(df['Date'], df['Close'], label='종가', color='black', linewidth=1, alpha=0.7)  
# [빈칸] 종가 데이터를 플롯하는 코드 작성

# 이동 평균선 그래프
plt.plot(df['Date'], df['MA5'], label='5일 이동 평균선', color='blue', linewidth=1)  
# [빈칸] 5일 이동 평균선을 플롯하는 코드 작성

plt.plot(df['Date'], df['MA20'], label='20일 이동 평균선', color='orange', linewidth=1)  
# [빈칸] 20일 이동 평균선을 플롯하는 코드 작성

plt.plot(df['Date'], df['MA60'], label='60일 이동 평균선', color='green', linewidth=1)  
# [빈칸] 60일 이동 평균선을 플롯하는 코드 작성

# 골든크로스 지점 강조
plt.scatter(golden_cross['Date'], golden_cross['Close'], color='red', label='골든크로스', marker='^', s=100)  
# [빈칸] 골든크로스 지점을 강조하는 코드 작성

# 데드크로스 지점 강조
plt.scatter(death_cross['Date'], death_cross['Close'], color='purple', label='데드크로스', marker='v', s=100)  
# [빈칸] 데드크로스 지점을 강조하는 코드 작성

# 그래프 제목 및 라벨 설정
plt.title('주식 데이터의 이동 평균선 분석 및 특정 구간 강조')
plt.xlabel('날짜')  # [빈칸] X축 라벨 설정
plt.ylabel('주가')  # [빈칸] Y축 라벨 설정
plt.legend()
plt.grid(True)

# 6. 결과 해석
# 1. 이동 평균선을 통해 주가의 흐름을 파악할 수 있다.
# 2. 5일 이동 평균선이 20일 이동 평균선을 상향 돌파하는 경우(골든크로스)는 
#    상승 신호로 해석된다.
# 3. 반대로, 5일 이동 평균선이 20일 이동 평균선을 하향 돌파하는 경우(데드크로스)는 
#    하락 신호로 해석된다.
# 4. shift(1) 함수는 이전 날의 데이터를 가져와서 현재와 비교하는 역할을 한다.
# 5. 이를 활용하면 "오늘의 이동 평균선이 교차했는지"를 정확히 판단할 수 있다.
# 6. 장기 이동 평균선(60일)을 함께 분석하면 
#    주가의 장기적인 방향성을 이해하는 데 도움을 줄 수 있다.
# 7. 특정 구간에서의 교차점(골든크로스, 데드크로스)을 강조하여 
#    투자 의사 결정을 지원할 수 있다.

# 그래프 출력
plt.show()  # [빈칸] 그래프를 출력하는 코드 작성
```

```python
# 2. 날짜 변환 및 연-월 데이터 생성
df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m")  # 'Date' 컬럼을 datetime 형식으로 변환
df
```

```python
# 연-월(YYYY-MM) 형식으로 변환하여 새로운 컬럼 생성
df["YearMonth"] = df["Date"].dt.to_period("M")
df
```

```python
# 3. 월별 매출 집계
monthly_sales = df.groupby("YearMonth")["Revenue"].sum().reset_index()  # 'YearMonth' 기준으로 매출 합산
monthly_sales["YearMonth"] = monthly_sales["YearMonth"].astype("str")  # 문자열로 변환하여 시각화 오류 방지
monthly_sales
```

```python
# 4. 제품별 총 판매량 분석
product_sales = df.groupby("Product")["Quantity"].sum().reset_index()  # 'Product'별로 총 판매량 합산
product_sales
```

```python
# 5. 가장 많이 팔린 제품 상위 10개 선정
top_products = product_sales.sort_values(by="Quantity", ascending=False).head(10)  # 판매량 기준으로 내림차순 정렬 후 상위 10개 선택
top_products
```

```python
# 6. 시각화 - 월별 매출 변화
plt.figure(figsize=(12, 6))  # 그래프 크기 설정
sns.lineplot(data=monthly_sales, x="YearMonth", y="Revenue", marker="o")  # 선 그래프(lineplot) 생성
plt.xticks(rotation=45)  # x축 라벨을 기울여 가독성 향상
plt.title("월별 매출 변화")  # 그래프 제목 설정
plt.xlabel("연-월")  # x축 레이블 설정
plt.ylabel("매출")  # y축 레이블 설정
plt.grid()  # 격자 추가 (grid)
plt.show()  # 그래프 출력
```

```python
# 7. 시각화 - 가장 많이 팔린 제품
plt.figure(figsize=(12, 6))  # 그래프 크기 설정
sns.barplot(data=top_products, x="Product", y="Quantity")  # 바 그래프(barplot) 생성
plt.title("가장 많이 팔린 제품 상위 10개")  # 그래프 제목 설정
plt.xlabel("제품")  # x축 레이블 설정
plt.ylabel("판매량")  # y축 레이블 설정
plt.xticks(rotation=45)  # x축 라벨 45도 회전
plt.show()  # 그래프 출력
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'

# 1. 데이터 불러오기 및 기본 탐색
# 주어진 영화 리뷰 데이터 파일을 불러온다.
file_path = "data/movie_reviews.csv"  # 실제 데이터 파일 경로
df = pd.read_csv(file_path)  # [빈칸] 데이터를 불러오는 코드 작성
df
```

```python
# 2. 데이터 기본 정보 확인
# - 데이터의 행과 열 개수, 결측치 여부, 데이터 타입 등을 확인합니다.
print("데이터 정보:")
df.info()
```

```python
# - 데이터의 일부 샘플을 출력하여 구조를 파악합니다.
print("\n데이터 샘플:")
df.sample(10)
```

```python
# 3. 리뷰 길이 계산 및 새로운 컬럼 추가
# - 각 리뷰의 문자 길이를 계산하여 새로운 컬럼 'Review Length'를 생성합니다.
# apply 안에는 함수를 작성
df["Review Length"] = df["Review"].apply(len)
df
```

```python
# 4. 평점과 리뷰 길이의 관계 시각화
# - 박스 플롯을 사용하여 평점별 리뷰 길이의 분포를 시각화합니다.
# - 이를 통해 특정 평점에서 리뷰 길이가 더 긴 경향이 있는지 확인할 수 있습니다.
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["Rating"], y=df["Review Length"])
plt.xlabel("Rating (평점)")
plt.ylabel("Review Length (리뷰 길이)")
plt.title("평점과 리뷰 길이의 관계")
plt.show()
```

```python
# 5. 리뷰 길이의 분포 확인
# - 히스토그램과 KDE(커널 밀도 추정)를 사용하여 리뷰 길이의 분포를 확인합니다.
# - 리뷰 길이의 평균과 편차를 시각적으로 파악할 수 있습니다.
plt.figure(figsize=(10, 6))
sns.histplot(df["Review Length"], bins=20, kde=True)  # 빈 개수와 KDE 적용 여부 지정
plt.xlabel("Review Length (리뷰 길이)")
plt.ylabel("Count (개수)")
plt.title("리뷰 길이 분포")
plt.show()
```

```python
# 6. 장문 리뷰 필터링 (상위 25% 이상의 길이를 가진 리뷰)
# - 전체 리뷰 길이 데이터에서 75번째 백분위수(Q3)를 계산합니다.
# - Q3 이상인 리뷰를 '장문 리뷰'로 간주하여 필터링합니다.
long_review_threshold = df["Review Length"].quantile(.75)  # 75% 백분위수
long_reviews = df[df["Review Length"] >= long_review_threshold]
print(f'전체 리뷰 길이 데이터에서 75번째 백분위수(Q3) : {long_review_threshold}')
long_reviews
```

```python
# 7. 필터링된 장문 리뷰 개수 및 샘플 출력
# - 장문 리뷰의 개수를 출력하여 얼마나 많은 리뷰가 해당하는지 확인합니다.
print(f"장문 리뷰 개수: {long_reviews.shape[0]}")

# - 일부 장문 리뷰 샘플을 출력하여 필터링이 잘 되었는지 확인합니다.
print("\n장문 리뷰 샘플:")
long_reviews.sample(10)
```
